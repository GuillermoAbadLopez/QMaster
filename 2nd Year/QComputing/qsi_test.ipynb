```python
import matplotlib.pyplot as plt
import numpy as np
from scipy.special import comb


def compute_perr(n, lambda1, lambda2):
    total = 0.0
    for k in range(n + 1):
        binom = comb(n, k, exact=False)
        term1 = ((1 - lambda1) ** k) * (lambda1 ** (n - k))
        term2 = ((1 - lambda2) ** k) * (lambda2 ** (n - k))
        total += binom * abs(term1 - term2)
    return 0.5 * (1 - 0.5 * total)


# Define the range of n values
ns = np.arange(1, 501, 10)

# Compute P_err for different lambda pairs
case4 = [compute_perr(n, 0.5, 0.5) for n in ns]
case5 = [compute_perr(n, 0.0, 0.001) for n in ns]
case6 = [compute_perr(n, 0.0, 0.01) for n in ns]
case2 = [compute_perr(n, 0.3, 0.4) for n in ns]
case1 = [compute_perr(n, 0.2, 0.5) for n in ns]
case3 = [compute_perr(n, 0.0, 1.0) for n in ns]

# Plot the results
plt.figure(figsize=(10, 6))

plt.plot(ns, case4, "^-", label="λ₁=0.5, λ₂=0.5, |λ₂-λ₁|=0")
plt.plot(ns, case5, "x-", label="λ₁=0.0, λ₂=0.001, |λ₂-λ₁|=0.001")
plt.plot(ns, case6, "v-", label="λ₁=0.0, λ₂=0.01, |λ₂-λ₁|=0.01")
plt.plot(ns, case2, "s-", label="λ₁=0.3, λ₂=0.4, |λ₂-λ₁|=0.1")
plt.plot(ns, case1, "o-", label="λ₁=0.2, λ₂=0.5, |λ₂-λ₁|=0.3")
plt.plot(ns, case3, "d-", label="λ₁=0.0, λ₂=1.0, |λ₂-λ₁|=1.0")
plt.xlabel("n")
plt.ylabel("$P_{err}$")
plt.title("Error Probability as a Function of n for Different λ₁ and λ₂")
plt.legend()
plt.grid(False)
plt.show()
```
